<h2>Model Training</h2>

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size=0.25, random_state=0)
```

```python
##Applying SMOTE for unbalanced dataset
```

```python
!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE(sampling_strategy='auto',random_state=42)
```

```python
X_train,y_train=smote.fit_resample(X_train,y_train)
```

```python
sns.countplot(y_train)
```

```python
#we've dealt with imbalanced dataset
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
treesmodel=ExtraTreesClassifier(n_estimators=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
from sklearn.metrics import accuracy_score,plot_confusion_matrix,classification_report
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='rbf')
```

```python
from sklearn.
```

```python
def evaluate(model):
    model.fit(X_train,y_train)
    preds=model.predict(X_test)
    print('Model accuracy is',accuracy_score(y_test,preds))
    plot_confusion_matrix(model,X_test,y_test)
    print(classification_report(y_test,preds))
```

```python
evaluate(logmodel)
```

```python
evaluate(treesmodel)
```

```python
evaluate(dt)
```

```python
evaluate(knn)
```

```python
evaluate(svc)
```
